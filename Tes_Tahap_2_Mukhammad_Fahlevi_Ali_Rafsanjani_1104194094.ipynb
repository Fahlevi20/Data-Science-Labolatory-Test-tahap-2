Loop

```python
# Terdapat 2 Loop Pada Python yaitu For Loop dan While Loop, Tugas kalian adalah
# dari range 0 sampai 30 ambil bilangan genapnya saja lalu simpan di variable bilangan genap
# Menggunakan For loop Dan While Loop

bilanganGenapForLoop = []
bilanganGenapWhileLoop = []
Bilangan = 30

## For Loop Section  ##
# Tulis Kodingan Disini Untuk For Loop
for x in range(0,Bilangan):
  if x%2==0:
    bilanganGenapForLoop.append(x)

##                   ##
## While loop Section ##
n=0
while n<Bilangan:
  if n%2==0:
    bilanganGenapWhileLoop.append(n)
  n+=1
# Tulis Kodingan disini Untuk while Loop

##                    ##
print(bilanganGenapForLoop)
print(bilanganGenapWhileLoop)
```
